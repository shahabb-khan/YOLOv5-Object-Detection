```python
import torch

# Load pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # 'yolov5s' is the small version
```

```python
import cv2

# Load an image
image_path = '/content/aaa.jpg'
image = cv2.imread(image_path)  # Load image in BGR format
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
```

```python
# Perform object detection
results = model(image)

# Display results
results.show()  # Opens a window with the detected objects
```

```python
from google.colab.patches import cv2_imshow # Import the alternative function

# ... your existing code ...

# Display the image with detections using cv2_imshow instead of cv2.imshow
cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Save the output image
output_path = 'output_image.jpg'
cv2.imwrite(output_path, cv2.cvtColor(image, cv2.COLOR_RGB2BGR))
```
